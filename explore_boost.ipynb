```python
import euclidemu2 as ee2
import numpy as np
import matplotlib.pyplot as plt

cosmo_par = {
    'As': 2.1e-09,
    'ns': 0.966,
    'Omb': 0.04,
    'Omm': 0.3,
    'h': 0.68,
    'mnu': 0.15,
    'w': -1.0,
    'wa': 0.0
}

cosmo_par_alt = cosmo_par.copy()
cosmo_par_alt['w'] = -0.7
cosmo_par_alt['wa'] = -0.

redshifts = [0, 2, 4, 6, 8, 10]

k, b = ee2.get_boost(cosmo_par, redshifts)
_, b_alt = ee2.get_boost(cosmo_par_alt, redshifts)

# --- Boost plot ---
plt.figure(figsize=(8, 6))
for i, z in enumerate(redshifts):
    plt.loglog(k, b[i], label=f"z={z} (w=-1)")
    plt.loglog(k, b_alt[i], linestyle='dotted', label=f"z={z} (w=-0.7, wa=0)")
plt.xlabel(r'$k\ [h/\mathrm{Mpc}]$')
plt.ylabel('boost')
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.3)
plt.tight_layout()
plt.title("Boost Factor for Two Dark Energy Models")
plt.show()

# --- Ratio plot ---
plt.figure(figsize=(8, 6))
for i, z in enumerate(redshifts):
    ratio = b_alt[i] / b[i]
    plt.semilogx(k, ratio, label=f"z={z}")
plt.axhline(1.0, color='black', linestyle='--', linewidth=1)
plt.xlabel(r'$k\ [h/\mathrm{Mpc}]$')
plt.ylabel(r'boost ratio: $(w=-0.5,\ wa=-0.7) \,/\, (w=-1,\ wa=0)$')
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.3)
plt.tight_layout()
plt.title("Boost Ratio Between Dark Energy Models")
plt.show()
```
